```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('coastal_news_geocoded.csv')
```

```python
df.head()
```

```python
fig = px.scatter_map(df,
                     lat="latitude",
                     lon="longitude",
                     color="category",
                     hover_name='headlines',
                     hover_data={
                         'headlines': False,
                         'date': True,
                         'location': True,
                         'latitude': False,
                         'longitude': False
                     })

fig.update_layout(
    autosize=True,
    map=dict(
        center=dict(lat=-30, lon=24),
        zoom=4,
        style='open-street-map',
        
    ),
    height=600
)
fig.show()
```
