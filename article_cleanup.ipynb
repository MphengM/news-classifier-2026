```python
import pandas as pd
import plotly.express as px
```

```python
# Load data without a header -see original file
df = pd.read_excel("DSIDE_data_oceans_and_coasts_news_articles - Sheet1.ods", header=None)
```

```python
df.iloc[8]
```

```python
# Keep only useful columns and give them useful names
df_clean = df[[0, 1, 2, 5, 6, 19]].copy()
df_clean.columns = ["headlines", "publication_date", "url", "location", "event_date", "category"]

df_clean.head(10)
```

```python
# Convert dates
df_clean['publication_date'] = pd.to_datetime(df_clean['publication_date'], dayfirst=True, errors='coerce')
df_clean['event_date'] = pd.to_datetime(df_clean['event_date'], dayfirst=True, errors='coerce')
```

```python
df_clean.head(10)
```

Publication or event dates couldn't be parsed for rows 0,1 , and 4. Row 6, 7, and 8 have the name of the publication instead of the category. Things aren't looking good for my already tiny dataset.

```python
# How many have at least ONE valid date?
has_pub_date = df_clean['publication_date'].notna()
has_event_date = df_clean['event_date'].notna()
has_any_date = has_pub_date | has_event_date

print(f"Articles with at least one date: {has_any_date.sum()} / {len(df_clean)}")

# How many have valid categories (not publication names)?
# Let's see all unique categories
print(f"\nAll categories:\n{df_clean['category'].value_counts()}")
```

```python
# Fill missing event dates with publication dates
df_clean['date'] = df_clean['event_date'].fillna(df_clean['publication_date'])
```

```python
df_clean.head(10)
```

```python
# Remove the 3 rows with publication names as categories
df_clean = df_clean[~df_clean['category'].isin(['bdlive', 'iol', 'ewn'])]
```

Final check that everything is fine now:

```python
print(f"Clean dataset: {len(df_clean)} articles")
print(df_clean['category'].value_counts())
```
